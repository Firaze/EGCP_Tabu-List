```python
def tabu_search(s0,t_max,verbose=False):
    if (verbose):
        print("Stato iniziale:  ",count_percolor(s0))
    best=list(s0)
    base=baseline_evaluate(best)
    tabulist=[]
    tabu_limit=10
    i=0
    fit_best=9999
    etime=0
    start_time=0
    while etime/60<=t_max and base.sum()>0:
        start_time=time.time()
        cand_c=np.random.permutation(list(np.where(base>0))[0])[0]
        sNeighborhood=getNeighbors(best,cand_c)
        np.random.shuffle(sNeighborhood)
        best_cand=0
        fit_bcand=999999
        bcand_c=0
        for cand in sNeighborhood:
            diff=np.abs(np.array(cand)-np.array(best))
            if(len(diff[diff>0])>1):
                cand_c1=np.argmax(diff)
                cand_c2=np.argmin(diff)
            else:
                cand_c1=np.argmax(diff)
                cand_c2=-1
            if (([best,cand] in tabulist)==False):
                fit_cand=evaluate(cand,cand_c1,cand_c2)
                bcand_c=int(cand_c)
            else:
                continue
            if (fit_cand<=fit_bcand):
                best_cand=cand
                fit_bcand=fit_cand
                bcand_c1=np.argmax(diff)
                bcand_c2=np.argmin(diff)
            
        tabulist.append([best,best_cand])
        best=best_cand
        fit_best=fit_bcand
        base=baseline_evaluate(best)
        if (verbose):
            print("evaluate:",base.sum())

        if (len(tabulist)>=tabu_limit):
            tabulist.pop(0)
        i+=1
        etime+=time.time()-start_time
    print("Iterazioni:  ",i+1)
    print("Tempo(s): ",int(etime))
    print("-----------------------------")
    return s0,best
```
